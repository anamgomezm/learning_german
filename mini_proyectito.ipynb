- Función <span style="color:purple">**buscador(palabra)**</span>:

```python
def buscador(palabra):
    """ Para buscar si ya tenemos la palabra en alemán o español"""
    df = pd.read_csv("vocabulario.csv", delimiter=";",header=0)
    try:
        filtro_aleman = df[df['Alemán'].str.contains(palabra, case=False)]
        if not filtro_aleman.empty:
            print(filtro_aleman, "\n")
    except:
        pass
    
    try:
        filtro_español = df[df['Español'].str.contains(palabra, case=False)]
        if not filtro_español.empty:
            print(filtro_español, "\n")
    except:
        pass
```

- Función <span style="color:purple">**repasar()**</span>:

Para esta función, también creé otra función auxiliar llamada *datos_seguimiento(si_no)* que servirá para más adelante representar el progreso.

*datos_seguimiento* almacena en *datos_seguimiento.csv* la fecha del día de hoy y si has acertado o no. Lo peculiar es que almacena solo los datos del día actual. Si los datos almacenados son de otro día, almacenará en *evolucion.csv* la fecha, el porcentaje de aciertos y el número de intentos de ese día para luego empezar a almacenar desde 0 en *datos_seguimiento.csv* los nuevos datos día actual.

La idea es no tener un fichero *datos_seguimiento.csv* enorme, sino obtener los datos que necesitamos almacenando directamente el resultado del tratamiento de esos datos en *evolucion.csv*. 

```python
from datetime import date
import pandas as pd

def datos_seguimiento(si_no):
    """ Esta función auxiliar nos servirá para representar el progreso más adelante. Creo que tendré que 
    implementar unas pruebas más, porque si el fichero se quedad vacío, tenemos un problema. Aunque no
    debería de quedarse vacío. Lo probaré durante un tiempo.
    """
    df = pd.read_csv("datos_seguimiento.csv", delimiter=";",header=0)
    hoy = str(date.today())
    
    if df.loc[0]["Día"] == hoy: # comprobamos si los datos son del día de hoy. En ese caso, añade 
        df = df.append({'Día' : hoy , 'Acierto' : si_no}, ignore_index=True)
        df.to_csv(r'/home/ana/Documents/curso_python/cuadernos/datos_seguimiento.csv',sep = ';',header=True, index=False)
 
    else: # si no son del día de hoy:
       
        # Antes de vaciar el fichero, recogemos los datos que nos interesan y los exportamos a otro fichero
        try:
            dia = df.loc[0]["Día"]
            p_aciertos = round(df.Acierto.str.count("Sí").sum()/len(df.index),3)
            intentos = len(df.index)

            # Los exportamos
            d = pd.read_csv("evolucion.csv", delimiter=";",header=0)
            d = d.append({'Día' : dia , 'Porcentaje de aciertos' : p_aciertos, 'Intentos' : intentos} , ignore_index=True)
            d.to_csv(r'/home/ana/Documents/curso_python/cuadernos/evolucion.csv',sep = ';',header=True, index=False)
        except:
            pass
        # A continuación, vaciamos y empezamos a rellenar otra vez el fichero datos_seguimiento
        
        df2 = pd.DataFrame([[hoy, si_no]])
        df2.columns = ["Día", "Acierto"]
        df2.to_csv(r'/home/ana/Documents/curso_python/cuadernos/datos_seguimiento.csv',sep = ';',header=True, index=False)
```

```python
def repasar():
    """ Esta función maravillosa te pregunta el vocabulario """
    df_pro = pd.read_csv("vocabulario.csv", delimiter=";",header=0)
    df_pro["Alemán"] = df_pro["Alemán"].apply(lambda x: x.strip()) 
    sample = df_pro.sample(1)
    print(sample["Español"].to_string(index=False))
    s = sample["Alemán"].to_string(index=False)
    x = input("auf Deutsch?: ",)
    if x:
        if x in s and len(x) >=3: # me di cuenta de que si escribo solo la primera letra y coincide con
            # la palabra, la da por válida
            print("Muy bien! Efectivamente, la palabra era: ", s)
            datos_seguimiento("Sí")
        else:
            print("Casi! La palabra era: ",s)
            datos_seguimiento("No")
    else:
        print("Vaya, la palabra era: ",s)
```

- Funciones <span style="color:purple">**dic_español(palabra)**</span> y <span style="color:purple">**dic_aleman(palabra)**</span>:

Fue necesario crear una función auxiliar *internet(url,palabra)*, que se encarga del web scraping.

```python
from selenium import webdriver
import time

def internet(url,palabra):
    
    """
    La función internet(url,palabra) requerirá del url y de la palabra a consultar.
    El motivo de hacerlo con el url es porque me di cuenta de que según si se busca de español a alemán o 
    al revés, la url es igual pero cambiando el orden, y la última palabra es la consultada. Entonces 
    pensé que para qué iba a hacerlo desde el html si así es más fácil
    """
    
    driver = webdriver.Firefox() # para abrir el firefox
    driver.get(url+palabra) # consigue la pág web exacta
    time.sleep(1) # aunque dijo el profe que esto no es necesario, pero para un futuro, que no se me olvide
    # Ahora, tenemos que clicar sobre el botón de aceptar las cookies:
    driver.find_element_by_css_selector('button[class="sc-bwzfXH jlyVur"]').click() # esta fantasía
    # la encontré aquí:
    # https://stackoverflow.com/questions/50334940/python-selenium-hover-and-click-not-working-on-webelement-whose-class-changes
    time.sleep(1) 

# Definimos las funciones dic:

def dic_español(español):
    """ Para consultar PONS en español"""
    internet("https://es.pons.com/traducción/español-alemán/",español)

def dic_aleman(aleman):
    """ Para consultar PONS en alemán"""
    internet("https://es.pons.com/traducción/alemán-español/",aleman)  
```

Como quiero usar este programa para estudiar en mi día a día, pues he creado también una función _motivacion_, porque nunca viene mal jajaja

```python
#from dateutil.relativedelta import relativedelta
from datetime import datetime

# Quería probar algún decorador
def our_decorator(func):
    def function_wrapper():
        func()
        texto = ("\nMan darf nie an die ganze Straße auf einmal denken, verstehst du? \
Man muß nur an den nächsten Schritt denken, an den nächsten Atemzug, \
an den nächsten Besenstrich. Und immer wieder nur an den nächsten. \
Dann macht es Freude; das ist wichtig, dann macht man seine Sache gut. \
Und so soll es sein. Auf einmal merkt man, daß man Schritt für Schritt die \
ganze Straße gemacht hat. Man hat gar nicht gemerkt wie, und man ist nicht \
außer Puste. - Momo, Michael Ende\n" )
        print("\nY ahora, un mini texto: \n",texto, "\n:D")
    return function_wrapper


@our_decorator
def motivacion():
    """Esta función se explica por sí misma"""
    examen = datetime.strptime("2021/06/02", "%Y/%m/%d")
    diff = examen - datetime.now()
    if diff.days > 60:
        print("Venga, menos drama que te quedan %d días para aprobar el examen ;)" %diff.days)
        print("Así que motivaciooooooon a topeeeee\n")
    if diff.days < 60 and diff.days > 30:
        print("Ánimo que queda menos!!!")
    if diff.days < 30 and diff.days >10:
        print("Que no panda el cúnico, lo tienes controlado")
    if diff.days < 10:
        print("UUUUUH ahora sí, pánico absolutoooo O: \n Es bromi, ya está hecho!!!")
```
